```python
import pandas as pd
```

```python
data = pd.read_csv("datasets/Existing_Commercial_Buildings_Energy_Performance_Ordinance_Report.csv")
```

## Which postal code in SF has the most energy efficient buildings

```python
data.columns
```

```python
energy_zip = pd.concat([data['2014 ENERGY STAR Score'], data['Postal Code']], axis=1)
```

#### page 185 in Python for Data Analysis tells you how to concatonate series objects
#### new remove N/As

```python
cleaned_energy_zip = energy_zip.dropna()
```

```python
cleaned_energy_zip.count()
```

```python
cleaned_energy_zip
```

```python
zip_grouped = cleaned_energy_zip['2014 ENERGY STAR Score'].groupby(cleaned_energy_zip['Postal Code'])
```

```python
zip_grouped.mean()
```

```python
zip_mean = zip_grouped.mean()
```

```python
zip_mean.to_csv('energy_zip_data.csv')
```
